### Part A

```python
import numpy as np
import matplotlib.pyplot as plt

# Distribution parameters for mode A (high blood pressure)
mean_A = 140
std_dev_A = 15

# Distribution parameters for mode A (normal blood pressure)
mean_B = 120
std_dev_B = 10

# Create the x data to plot the distribution
x = np.linspace(80, 200, 400)  # Range of x from 80 to 200

# Calculate the probability density for both modes
pdf_A = (1 / (np.sqrt(2 * np.pi * std_dev_A**2))) * np.exp(-((x - mean_A)**2) / (2 * std_dev_A**2))
pdf_B = (1 / (np.sqrt(2 * np.pi * std_dev_B**2))) * np.exp(-((x - mean_B)**2) / (2 * std_dev_B**2))

# Draw distributions
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_A, label='Condition A (Hypertension)', color='blue')
plt.plot(x, pdf_B, label='Condition B (Normotension)', color='red')
plt.title('Probability Density Functions of Blood Pressure')
plt.xlabel('Systolic Blood Pressure (mmHg)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```

### Part B

```python
import numpy as np
from scipy.stats import norm

# Parameters for distribution for Condition A (Hypertension)
mean_A = 140  # Mean blood pressure
std_dev_A = 15  # Standard deviation

# Parameters for distribution for Condition B (Normotension)
mean_B = 120  # Mean blood pressure
std_dev_B = 10  # Standard deviation

# Range for searching the threshold
x_values = np.linspace(80, 200, 400)  # Blood pressure from 80 to 200

# Finding the decision threshold
threshold = None
for x in x_values:
    p_A = norm.pdf(x, mean_A, std_dev_A)  # Probability density for Condition A
    p_B = norm.pdf(x, mean_B, std_dev_B)  # Probability density for Condition B
    if (abs(p_A - p_B)/(p_A+p_B)) < 0.01:  # If the densities are close
        threshold = x
        break

# Calculate misclassification probability for each condition
p_error_A = norm.cdf(threshold, mean_A, std_dev_A)  # Misclassification probability for Condition A
p_error_B = 1 - norm.cdf(threshold, mean_B, std_dev_B)  # Misclassification probability for Condition B

# Display results
print(f'Threshold:{threshold}')
print("Misclassification probability for Condition A:", p_error_A)
print("Misclassification probability for Condition B:", p_error_B)

# Check which condition is more prone to misclassification
if p_error_A > p_error_B:
    print("Condition A (Hypertension) is more prone to misclassification.")
else:
    print("Condition B (Normotension) is more prone to misclassification.")
```

```python
import numpy as np

# Number of data points
n = 100

# Generate random data for three variables x1, x2, x3
x1 = np.random.randn(n)
x2 = np.random.randn(n)
x3 = np.random.randn(n)

# Generate the output y based on the nonlinear relationship with x1, x2, and x3
y = (2 * x1**2 + 3 * x2 + x3) / (1 + x1 + x2**2)

# Combine x1, x2, and x3 into a matrix X
X = np.column_stack((x1, x2, x3))

# Formula: a = (X.T * X)^(-1) * X.T * y
a = np.linalg.inv(X.T @ X) @ (X.T @ y)

# Calculate predicted values for y based on the estimated parameters
y_pred = X @ a

# Compute the Mean Squared Error (MSE) between the actual and predicted y values
MSE = np.mean((y - y_pred) ** 2)

# Display the estimated parameters
print('Estimated parameters:')
print(a)

# Display the Mean Squared Error
print('Mean Squared Error:')
print(MSE)
```

```python
import numpy as np

# Number of data points
n = 100

# Generate random data for three variables x1, x2, x3
x1 = np.random.randn(n)
x2 = np.random.randn(n)
x3 = np.random.randn(n)

# Generate the output y based on the nonlinear relationship with x1, x2, and x3
y = (2 * x1**2 + 3 * x2 + x3) / (1 + x1 + x2**2)

# Create the design matrix for the quadratic model
X_quad = np.column_stack((x1**2, x2**2, x3, x1 * x2))

# Estimate parameters 'a_quad' using the Ordinary Least Squares (OLS) method
a_quad = np.linalg.inv(X_quad.T @ X_quad) @ (X_quad.T @ y)

# Calculate predicted values for y based on the estimated parameters for the quadratic model
y_pred_quad = X_quad @ a_quad

# Compute the Mean Squared Error (MSE) between the actual and predicted y values
MSE_quad = np.mean((y - y_pred_quad) ** 2)

# Display the estimated parameters for the quadratic model
print('Estimated parameters for quadratic model:')
print(f'a1:{a_quad[0]} , a2:{a_quad[1]} , a3:{a_quad[2]} , a4:{a_quad[3]}')

# Display the Mean Squared Error for the quadratic model
print('Mean Squared Error for quadratic model:')
print(MSE_quad)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of sample points
n_points = 1000

# Generate Gaussian white noise (mean 0, variance 1)
np.random.seed(0)  # For reproducibility of results
x = np.random.normal(0, 1, n_points)

# Initialize y[n] (with two initial values set to zero)
y = np.zeros(n_points)

# Calculate the system's output using the equation y[n]
for n in range(2, n_points):
    y[n] = 0.5 * y[n-1] - 0.1 * y[n-2] + x[n]

# Plotting input and output
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(x, label='Input (Gaussian Noise)')
plt.title('Input: Gaussian White Noise')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(y, label='Output', color='orange')
plt.title('Output of the System')
plt.legend()

plt.tight_layout()
plt.show()
```

## OPTIONAL
The following code addresses how we can determine the order of an AR model using only the input and output data.

```python
import numpy as np
from statsmodels.tsa.ar_model import AutoReg

# Generate random input and output data (blood pressure and heart rate)
np.random.seed(0)
length = 100  # Length of the data
x = np.random.normal(0, 1, length)  # Blood pressure
y = np.zeros(length)  # Heart rate

# Generate y based on hypothetical model
a1, a2 = 0.5, 0.3  # Parameters of the hypothetical model
b1, b2 = 0.7, 0.2

for t in range(2, length):
    y[t] = a1 * y[t-1] + a2 * y[t-2] + b1 * x[t-1] + b2 * x[t-2] + np.random.normal(0, 0.1)

# Function to estimate AR model and calculate AIC and BIC
def evaluate_ar_model(y, order):
    model = AutoReg(y, lags=order)  # Create AutoReg model with specified lags
    model_fit = model.fit()  # Fit the model to the data
    return model_fit.aic, model_fit.bic  # Return AIC and BIC values

# Check different orders
max_order = 10  # Maximum order to evaluate
aic_values = []  # List to store AIC values
bic_values = []  # List to store BIC values

for order in range(1, max_order + 1):
    aic, bic = evaluate_ar_model(y, order)  # Evaluate AR model for the current order
    aic_values.append(aic)  # Store AIC value
    bic_values.append(bic)  # Store BIC value
    print(f"Order: {order}, AIC: {aic}, BIC: {bic}")  # Print order, AIC, and BIC

# Find the best order based on AIC and BIC
best_aic_order = np.argmin(aic_values) + 1  # Find order with minimum AIC
best_bic_order = np.argmin(bic_values) + 1  # Find order with minimum BIC

print(f"\nBest order based on AIC: {best_aic_order}")  # Print best order based on AIC
print(f"Best order based on BIC: {best_bic_order}")  # Print best order based on BIC
```
